```
import pandas as pd
fileName = 'natural.csv'
samplesN = 500
df = pd.read_csv(fileName)
df
```

```
import plotly.graph_objects as go

def plotNumberOfGen(n=None):
    maxGen = n 
    if n == None:
        maxGen = max(df['geração'].tolist())
    x = []
    y = []
    for i in range(maxGen):
        x.append(i+1)
        y.append(len(df[df['geração'] == i+1]))

    fig = go.Figure(data=go.Scatter(x=x, y=y))
    fig.show()
    
```

```
plotNumberOfGen(10)
```

```
plotNumberOfGen()
```

```
def plotMeanDistance(n=None):
    maxGen = n 
    if n == None:
        maxGen = max(df['geração'].tolist())
    x = []
    y = []
    for i in range(maxGen):
        x.append(i+1)
        dfa = df[df['geração'] == i+1]
        y.append(dfa['meanDistance'].iloc[0])

    fig = go.Figure(data=go.Scatter(x=x, y=y))
    fig.show()
```

```
plotMeanDistance()
```

```
df1 = pd.read_csv('natural.csv')
df2 = pd.read_csv('elitismo.csv')
df3 = pd.read_csv('tournament.csv')
df4 = pd.read_csv('roleta.csv')
```

```
def plotMeanDistanceAll(DFlist,Names=None ,n=None):
    maxGen = n 
    dfNum = len(DFlist)
    if n == None:
        maxGen = float("inf")
        for i in range(dfNum):
            maxGen= min(maxGen, max(DFlist[i]['geração'].tolist()))

    
    x = []
    y = [[] for i in range(dfNum)]
    for i in range(maxGen):
        x.append(i+1)
        for j in range(dfNum):
            dfaux = DFlist[j]
            dfaux =  dfaux[dfaux['geração'] == i+1]
            y[j].append(dfaux['meanDistance'].iloc[0])
    
    fig = go.Figure()
    for i in range(dfNum):
        if Names == None:
            fig.add_trace(go.Scatter(x=x, y=y[i],
                        mode='lines+markers',
                        name='df ' + str(i)))
        else:
            fig.add_trace(go.Scatter(x=x, y=y[i],
                        mode='lines+markers',
                        name=Names[i]))
    fig.show()
```

```
plotMeanDistanceAll([df1],["natural"])
plotMeanDistanceAll([df2],["elitismo"])
plotMeanDistanceAll([df3],["torneio de 2"])
plotMeanDistanceAll([df4],["roleta"])
```

```
plotMeanDistanceAll([df1, df2, df3, df4],["natural", "elitismo", "torneio de 2", "roleta"])
```
